```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
```

```python
# import data
traffic_2019= pd.read_csv("C:/Ironhack/Week9-FinalProject/traffic_2019_final")
```

```python
# drop unwanted columns
traffic_2019= traffic_2019.drop(columns=['Unnamed: 0.1','Unnamed: 0'])
```

```python
traffic_2019.head()
```

```python
traffic_2019.dtypes
```

```python
# make datetime column
traffic_2019['datetime'] = pd.to_datetime(traffic_2019['datetime'])
traffic_2019['hour'] = traffic_2019['datetime'].dt.hour
traffic_2019['date'] = traffic_2019['datetime'].dt.date
```

```python
traffic_2019
```

```python
# Group by 'id', 'date', and 'datetime', and aggregate numerical columns
traffic_hourly = traffic_2019.groupby(['id', 'tipo_elem', 'day', 'month', 'year','hour','date', 'district', 'Nombre,,,,']).agg({
    'intensity': 'sum',
    'occupancy': 'mean',
    'load': 'mean',
    'velocity': 'mean',
    'integration_period': 'mean'
}).reset_index()
```

```python
traffic_hourly
```

```python
traffic_hourly.dtypes
```

```python
# plot graphs for hourly data
import matplotlib.pyplot as plt

# Time Series Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='intensity', data=traffic_hourly, hue='district')
plt.title('Intensity Over Time')
plt.xlabel('Date')
plt.ylabel('Intensity')
plt.show()

# Time Series Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='intensity', data=traffic_hourly)
plt.title('Intensity Over Time')
plt.xlabel('Date')
plt.ylabel('Intensity')
plt.show()

# District-wise Comparison (Bar Chart)
plt.figure(figsize=(10, 6))
district_comparison = traffic_hourly.groupby('district')[['intensity']].mean()
district_comparison.plot(kind='bar', stacked=True)
plt.title('District-wise Comparison')
plt.xlabel('District')
plt.ylabel('Average Value')
plt.show()

# Correlation Matrix
correlation_matrix = traffic_hourly[['intensity', 'occupancy', 'load', 'velocity']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Seasonal Trends
monthly_trends = traffic_hourly.groupby('month')[['intensity']].mean()
plt.figure(figsize=(10, 6))
monthly_trends.plot(marker='o')
plt.title('Seasonal Trends')
plt.xlabel('Month')
plt.ylabel('Average Value')
plt.show()
```

```python
traffic_hourly_copy=traffic_hourly.copy()
```

```python
# make datetime column
traffic_hourly_copy['date'] = pd.to_datetime(traffic_hourly_copy['date'])
```

```python
import pandas as pd

# Extract week number, day of the week, and sensor ID from the 'date' column
traffic_hourly_copy['week'] = traffic_hourly_copy['date'].dt.week
traffic_hourly_copy['day_of_week'] = traffic_hourly_copy['date'].dt.day_name()

# Group by 'id', 'week', and 'day_of_week' columns, calculate the mean traffic count
df_avg_daily_traffic_per_sensor_avg_week = traffic_hourly_copy.groupby(['id', 'week', 'day_of_week','hour']).agg({'intensity': 'mean'}).reset_index()

df_avg_daily_traffic_per_sensor_avg_week
```

```python
# Filter the dataframe for the specific sensor (id=3396)
df_sensor_3396 = df_avg_daily_traffic_per_sensor_avg_week[df_avg_daily_traffic_per_sensor_avg_week['id'] == 3396]

# Time Series Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='intensity', data=df_sensor_3396, hue='day_of_week')
plt.title('Intensity Over Time for Sensor 3396')
plt.xlabel('Hour of the Day')
plt.ylabel('Intensity')
plt.show()
```

```python
# Assuming df_avg_daily_traffic_per_sensor_avg_week is the DataFrame containing average data for all sensors

# Time Series Plot for Average Data of All Sensors
plt.figure(figsize=(12, 6))
sns.lineplot(x='hour', y='intensity', data=df_avg_daily_traffic_per_sensor_avg_week, hue='day_of_week')

# Customize x-axis ticks to show every hour
plt.xticks(range(24))

plt.title('Average Intensity Over Time for All Sensors')
plt.xlabel('Hour of the Day')
plt.ylabel('Intensity')
plt.show()
```

```python
df_avg_intensity_overall = df_avg_daily_traffic_per_sensor_avg_week.groupby('hour')['intensity'].mean().reset_index()
df_avg_intensity_overall
```

```python
# Group by 'id', 'day_of_week', and 'hour', calculate the mean intensity
df_avg_intensity_per_sensor = traffic_hourly_copy.groupby(['id', 'day_of_week', 'hour']).agg({'intensity': 'mean'}).reset_index()

# Group by 'id' and 'hour', calculate the overall mean intensity
df_avg_intensity_overall = df_avg_intensity_per_sensor.groupby(['id', 'hour']).agg({'intensity': 'mean'}).reset_index()

# Merge the per-sensor average with the overall average
df_avg_week_intensity_per_sensor = pd.merge(df_avg_intensity_per_sensor, df_avg_intensity_overall, on=['id', 'hour'], suffixes=('_sensor', '_overall'))
```

```python
df_avg_week_intensity_per_sensor
```

```python
# Create a mapping DataFrame with unique sensor IDs and their corresponding latitude and longitude
sensor_mapping = traffic_2019[['id', 'latitude', 'longitude']].drop_duplicates()

# Merge the mapping DataFrame with df_avg_daily_traffic_per_sensor_avg_week
df_avg_week_intensity_per_sensor = pd.merge(df_avg_week_intensity_per_sensor, sensor_mapping, on='id')
```
